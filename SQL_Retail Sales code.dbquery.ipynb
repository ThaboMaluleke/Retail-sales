```sql
select * from `retail`.`sales`.`retail_sales_dataset_1` limit 50;

--Display only the Transaction ID, Date, and Customer ID for all records

SELECT 
    `Transaction ID`,
    `Date`,
    `Customer ID`
FROM retail.sales.retail_sales_dataset_1;


---Totaol number of tansactions

select count(distinct `Transaction ID`) as total_transactions,
count(distinct `customer id`) as total_customers
from retail.sales.retail_sales_dataset_1


--To SEE the product category

SELECT DISTINCT `Product Category`
FROM retail.sales.retail_sales_dataset_1

--DISPLAY SALES TRANSACTION PERFOMED BY MALE, UNDER ELECTRONICS

SELECT *
FROM retail.sales.retail_sales_dataset_1
WHERE `Gender` = 'Male' AND `Product Category` = 'Electronics'

--display ALL TRANSACTION MADE BY PEOPLE WITH THE AGFE RANGING FROM 20-40

SELECT *
FROM retail.sales.retail_sales_dataset_1
WHERE `Age` BETWEEN 20 AND 40
ORDER BY `AGE`
LIMIT 25;

--display ALL TRANSACTION MADE BY PEOPLE WITH THE AGFE RANGING FROM 20-40, AND PURCHASED ITEMS IN BEAUTY

SELECT *
FROM retail.sales.retail_sales_dataset_1
WHERE `Product Category` = "Beauty"
AND `Age` BETWEEN 20 AND 40

--display ALL TRANSACTION MADE BY PEOPLE WITH THE AGFE RANGING FROM 20-40, AND` `Age` BETWEEN 20 AND 40

LIMIT 25;
```
